# Airlines Passenger EDA
Exploratory Data Analysis to understand passenger satisfaction, delays, and loyalty behavior.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('airlines_data.csv')
df.head()
```

## Dataset Information

```python
df.info()
df.describe()
```

## Flight Class Distribution

```python
sns.countplot(x='Class', data=df)
plt.title('Distribution of Flight Classes')
plt.savefig('class_distribution.png')
plt.show()
```

## Departure Delay Distribution

```python
sns.histplot(df['Departure_Delay'], bins=30)
plt.title('Departure Delay Distribution')
plt.savefig('delay_distribution.png')
plt.show()
```

## Satisfaction by Flight Class

```python
sns.boxplot(x='Class', y='Satisfaction_Score', data=df)
plt.title('Customer Satisfaction by Class')
plt.savefig('satisfaction_boxplot.png')
plt.show()
```

## Correlation Heatmap

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.title('Correlation Matrix')
plt.show()
```

## Top Insights
1. Economy class passengers represent the largest group.
2. Business class passengers generally have higher satisfaction.
3. Higher delays often reduce satisfaction levels.
4. Loyal customers accumulate higher loyalty points.
5. Complaints increase when delays increase.
