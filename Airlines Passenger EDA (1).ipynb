```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Load Dataset

```python
df = pd.read_csv("airlines_data.csv")
df.head()
```

## Dataset Information

```python
df.info()
df.describe()
```

## Data Cleaning

```python
# Check missing values
df.isnull().sum()

# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Fill missing values with 0
df.fillna(0, inplace=True)

print("Dataset cleaned successfully")
```

## Flight Class Distribution

```python
plt.figure(figsize=(8,5))

sns.countplot(x='Class', data=df)

plt.title("Distribution of Flight Classes")

plt.xlabel("Flight Class")

plt.ylabel("Number of Passengers")

plt.savefig("class_distribution.png")

plt.show()
```

## Satisfaction by Flight Class

```python
plt.figure(figsize=(8,5))

sns.boxplot(x='Class', y='Satisfaction_Score', data=df)

plt.title("Customer Satisfaction by Flight Class")

plt.xlabel("Flight Class")

plt.ylabel("Satisfaction Score")

plt.savefig("satisfaction_boxplot.png")

plt.show()
```

## Correlation Heatmap

```python
plt.figure(figsize=(10,6))

# Select only numeric columns
correlation = df.select_dtypes(include=['number']).corr()

sns.heatmap(correlation, annot=True, cmap="coolwarm")

plt.title("Correlation Matrix")

plt.show()
```

## Delay vs Complaints Analysis

```python
plt.figure(figsize=(8,5))

sns.scatterplot(x='Departure_Delay', y='Complaints', data=df)

plt.title("Relationship Between Delay and Complaints")

plt.xlabel("Departure Delay")

plt.ylabel("Complaints")

plt.show()
```

# Summary

Airlines EDA Project Summary

This project performs exploratory data analysis on airline passenger data to understand customer satisfaction, delay patterns, and loyalty trends.

Key Insights:
1. Economy class dominates passenger volume.
